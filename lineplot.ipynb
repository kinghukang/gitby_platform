```python
plt.figure(figsize=(10,5))
sns.lineplot(x="Month",y="Passengers",data=df)
plt.title("month vs passengers")
plt.xlabel("month")
plt.ylabel("passengers")
plt.show()
```

```python
ax=sns.lineplot(x="Month",y="Passengers",data=df)
ax_ticket=ax.get_xticklabels()
plt.close()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x="Month",y="Passengers",data=df)
plt.title("month vs passengers")
plt.xlabel("month")
plt.xticks(range(23,144,24))
plt.ylabel("passengers")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x="Month",y="Passengers",data=df,marker=".", markersize=5, markerfacecolor="green")
plt.title("month vs passengers")
plt.xlabel("month")
plt.xticks(range(23,144,24))
plt.ylabel("passengers")
plt.show()
```

```python
plt.figure(figsize = (8, 3))
sns.lineplot(
    data = df,
    x = 'Month',
    y = 'Passengers',
    marker = 'o',
    markersize = 4,
    markerfacecolor = 'orange',
    lw = 1,
    ls = ':'
)
plt.title('Line plot of monthly number of passengers in the airline')
plt.xlabel('Month')
plt.xticks(range(23, 144, 24))
plt.ylabel('Passengers')
plt.ylim((0, 700))  # setting a range for the y-axis limits
plt.show()
```
