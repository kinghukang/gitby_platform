### Convert Date Columns

**Q1.5** - Convert all date-related columns to a standardised datetime format. **(2 marks)**
* **Hint** - Use datetime function

```python
customer["Birthday"] = pd.to_datetime(customer["Birthday"])
sales["Order Date"] = pd.to_datetime(sales["Order Date"])
sales["Delivery Date"] = pd.to_datetime(sales["Delivery Date"])
stores["Open Date"] = pd.to_datetime(stores["Open Date"])
exchange["Date"] = pd.to_datetime(exchange["Date"])
customer
```

### Transform Numerical Columns

**Q1.6** - Transform object-type columns containing numeric information into numerical data types for analysis. **(2 marks)**
* **Hint** - Use Replace function

```python
customer["Zip Code"] = customer["Zip Code"].replace(r"[^0-9]", "", regex=True)
customer["Zip Code"] = pd.to_numeric(customer["Zip Code"], errors="coerce")
customer["Zip Code"] = customer["Zip Code"].astype("Int64")
products["Unit Cost USD"] = (
    products["Unit Cost USD"]
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.strip()
    .astype(float)
)

products["Unit Price USD"] = (
    products["Unit Price USD"]
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.strip()
    .astype(float)
)

customer
```

### Join the Tables

```python
# Merge FX rates
sales_fx = sales.merge(
    exchange,
    left_on=['Order Date', 'Currency Code'],
    right_on=['Date', 'Currency'],
    how='left'
)

# Merge master data
sales_full = (
    sales_fx
    .merge(customer, on='CustomerKey', how='left')
    .merge(products, on='ProductKey', how='left')
    .merge(stores, on='StoreKey', how='left')
)
```

* A consolidated, analysis-ready dataset (sales_full) containing transactional, customer, product, store, and currency information in a single table.

### Normalise the Revenue

```python
# Create USD-normalised revenue
sales_full['Revenue_USD'] = (
    sales_full['Quantity'] *
    sales_full['Unit Price USD'] *
    sales_full['Exchange']
)
```

* Revenue was converted to USD to enable consistent cross-country financial analysis.
